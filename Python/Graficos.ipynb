```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Algoritmo 1

```python
#Numero de aristas
X = [2**16, 2**17, 2**18, 2**19, 2**20, 2 **21, 2**22, 2**23,2**24]
#Algoritmo 1: resultados
Y1 = [1.32045e+06, 1.34186e+06,1.40227e+06,1.51582e+06, 1.67122e+06, 1.88194e+06,
      2.30277e+06, 3.50573e+06, 6.16724e+06]
err1 = [10540.6, 12080.7,11547.8 ,23284.6,10224.5,18202.2,
       69028.8 , 97389.6 , 128313 ]
```

```python
plt.plot(X, Y1)
#Agregamos las etiquetas y añadimos una leyenda.
plt.errorbar(X,Y1, linestyle="None", yerr = err1, fmt="ob", capsize=5, ecolor="k")

plt.xlabel('N° de aristas')
plt.ylabel('Tiempo [µs].')
plt.title("Algoritmo 1: escala lineal")
plt.show()
```

```python
plt.xscale("log")
plt.plot(X, Y1)
#Agregamos las etiquetas y añadimos una leyenda.
plt.errorbar(X,Y1, linestyle="None", yerr = err1, fmt="ob", capsize=7, ecolor="k")
plt.xlabel('Cantidad de aristas')
plt.ylabel('Tiempo [µs].')
plt.title("Algoritmo 1: Escala log")
plt.show()
```

# COMPARACION DE LOS 3 ALGORITMOS


```python
plt.plot(X, Y2, label  = "Algoritmo 2")
plt.plot(X, Y1, label  = "Algoritmo 1")
plt.plot(X, Y3, label  = "Algoritmo 3 ")
plt.errorbar(X,Y3, linestyle="None", yerr = err3, fmt="ob", capsize=7, ecolor="k")
plt.errorbar(X,Y2, linestyle="None", yerr = err2, fmt="ob", capsize=7, ecolor="k")
plt.errorbar(X,Y1, linestyle="None", yerr = err1, fmt="ob", capsize=7, ecolor="k")

#Agregamos las etiquetas y añadimos una leyenda.
plt.xlabel('N° de aristas')
plt.ylabel('Tiempo [µs].')
plt.title("Comparación: escala lineal")
plt.legend()
plt.show()
```

# COMPARACION ERROR

```python
#Transformamos a porcentaje
i=0
e1= []
e2= []
e3= []
e4=[]
while i<9:
  e1.append(round(err1[i]*100/Y1[i], 3))
  e2.append(round(err2[i]*100/Y2[i], 3))
  e3.append(round(err3[i]*100/Y3[i], 3))
  e4.append(round(err3_2[i]*100/Y3[i], 3))
  i=i+1
 
```

```python
print(e1)
print(e2)
print(e3)
print(e4)
```

```python
plt.plot(X,e1, label = "Alg 1 con 50 iter")
plt.plot(X,e2, label = "Alg 2 con 50 iter")
plt.plot(X,e3, label = "Alg 3 con 50 iter")
plt.plot(X,e4, label = "Alg 1 con 500 iter")

#Agregamos las etiquetas y añadimos una leyenda.
plt.xlabel('N° de aristas')
plt.ylabel('Porcentaje de error')
plt.title("Comparación desviación estándar (lineal)")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.show()
```
